__Задание 8__
<p>Создайте два Series :
<li>s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])</li>
<li>s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])</li> Выполните сложение s1 + s2 . Объясните, почему в результате появляются NaN , и замените их на 0 .</p>
<p>Объяснение Индекс 'a' есть только в s1, в s2 отсутствует - результат NaN. Индекс 'e' есть только в s2, в s1 отсутствует - результат NaN. Pandas выполняет операцию по совпадающим индексам. Для отсутствующих значений результат становится NaN.</p> 

```python
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

print("s1:")
print(s1)
print("\ns2:")
print(s2)

s_sum = s1 + s2
s_sum_filled = s_sum.fillna(0)
print("\nПосле замены NaN на 0:")
print(s_sum_filled)
```

__Задание 9__
<p>Создайте Series из чисел [2, 4, 6, 8, 10] . Напишите код, который применяет к каждому элементу функцию вычисления квадратного корня с помощью .apply(np.sqrt) .</p>

```python

s9 = pd.Series([2, 4, 6, 8, 10])
print("Исходный Series:")
print(s9)

s9_sqrt = s9.apply(np.sqrt)
print("\nПосле применения np.sqrt:")
print(s9_sqrt)

s9_sqrt_alt = np.sqrt(s9)
print("\nАльтернативный способ (np.sqrt(s9)):")
print(s9_sqrt_alt)

def custom_sqrt(x):
    return x ** 0.5

s9_custom = s9.apply(custom_sqrt)
print("\nС применением пользовательской функции:")
print(s9_custom)
print()
```

__Задание 12__
<p>Создайте Series синдексами ['A', 'B', 'A', 'C', 'D', 'B'] изначениями [10, 20, 30, 40, 50, 60] . Проверьте, являются ли индексы уникальными. Если нет, сгруппируйте повторяющиеся индексы и сложите их значения.</p>

```python
s12 = pd.Series([10, 20, 30, 40, 50, 60], index=['A', 'B', 'A', 'C', 'D', 'B'])
print("Исходный Series с повторяющимися индексами:")
print(s12)

is_unique = s12.index.is_unique
print(f"\nИндексы уникальны? {is_unique}")

if not is_unique:
    duplicate_mask = s12.index.duplicated()
    print(f"Маска дубликатов (True - повтор):")
    print(duplicate_mask)
    
    print(f"\nИндексы, которые повторяются:")
    print(s12.index[s12.index.duplicated()].unique().tolist())
    
    print("\nГруппировка по индексам с суммированием значений:")
    s12_grouped = s12.groupby(s12.index).sum()
    print(s12_grouped)
    
    print("\nГруппировка с разными агрегациями:")
    print(s12.groupby(s12.index).agg(['sum', 'mean', 'count']))
    
    print(f"\nПосле группировки индексы уникальны? {s12_grouped.index.is_unique}")
```

__Задание 15__
<p>Создайте Series , где индексами будут даты с 1 по 30 марта 2024 года, а значениями - случайные числа от 50 до 150. Постройте график значений с помощью matplotlib . Добавьте заголовок, подписи осей и сетку.</p>

```python
import matplotlib.pyplot as plt
np.random.seed(42)
dates = pd.date_range('2024-03-01', '2024-03-30')
s15 = pd.Series(np.random.randint(50, 151, size=len(dates)), index=dates)

print("Создан временной ряд (1-30 марта 2024):")
print("Первые 5 элементов:")
print(s15.head())
print(f"\nРазмер: {len(s15)}")
print(f"Тип индекса: {type(s15.index)}")

fig, ax = plt.subplots(figsize=(14, 7))

ax.plot(s15.index, s15.values, marker='o', linewidth=2, 
        markersize=4, color='blue', label='Значение')

# Добавление горизонтальной линии среднего значения
mean_val = s15.mean()
ax.axhline(y=mean_val, color='red', linestyle='--', alpha=0.7, 
           label=f'Среднее ({mean_val:.1f})')

ax.fill_between(s15.index, s15.min(), s15.max(), alpha=0.1, color='gray')

# Оформление 
ax.set_title('Динамика значений за март 2024 года', fontsize=16, fontweight='bold')
ax.set_xlabel('Дата', fontsize=14)
ax.set_ylabel('Значение', fontsize=14)
ax.grid(True, alpha=0.3, linestyle='--')

# Настройка оси X
ax.set_xticks(s15.index[::3])
ax.set_xticklabels([d.strftime('%d.%m') for d in s15.index[::3]], rotation=45)

ax.legend(loc='upper left', fontsize=12)

# подписей значений
for i, (idx, val) in enumerate(s15.items()):
    if i % 5 == 0:  
        ax.annotate(f'{val}', (idx, val), textcoords="offset points", 
                    xytext=(0, 8), ha='center', fontsize=7, color='darkblue')

plt.tight_layout()
plt.show()
```

__Индивидуальное задание 4__
<p>Создайте CSV-файл water_usage.csv со следующими данными:</p>
<p>Дата,Потребление воды (м³)</p>
<p>2024-10-01,12000</p>
<p>2024-10-02,12500</p>
<p>2024-10-03,11800</p>
<p>2024-10-04,13000</p>
<p>2024-10-05,13500</p>

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_data = """Дата,Потребление воды (м³)
2024-10-01,12000
2024-10-02,12500
2024-10-03,11800
2024-10-04,13000
2024-10-05,13500
"""

with open('water_usage.csv', 'w', encoding='utf-8') as f:
    f.write(csv_data)

print("Файл 'water_usage.csv' создан")

df = pd.read_csv('water_usage.csv')
print("\nДанные из файла:")
print(df)

df['Дата'] = pd.to_datetime(df['Дата'])
df = df.set_index('Дата')

s = df['Потребление воды (м³)']
print("\nСозданный Series:")
print(s)

print("\nСтатистика:")
print(f"Среднее: {s.mean():.2f} м³")
print(f"Минимум: {s.min()} м³")
print(f"Максимум: {s.max()} м³")
print(f"Сумма: {s.sum()} м³")

plt.figure(figsize=(10, 5))
plt.plot(s.index, s.values, marker='o', linewidth=2, markersize=8, color='blue')
plt.title('Потребление воды', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Потребление (м³)', fontsize=12)
plt.grid(True, alpha=0.3)

for idx, val in s.items():
    plt.annotate(f'{val}', (idx, val), textcoords="offset points", xytext=(0, 10), ha='center')

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
